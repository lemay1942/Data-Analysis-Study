```python
with open('dp.html','wt', encoding='utf-8') as f:
    f.write(html)
```

```python
with open('dp.html', 'rt', encoding='utf-8') as f:
    html = f.read()
```

```python
soup = bs(html, 'html.parser')
```

```python
soup
```

```python
soup.find_all('span', class_='teams-navbar-module__team-link--name')
```

```python
title_list = []
for title in soup.find_all('span', class_='teams-navbar-module__team-link--name'):
    print(title.text)
    title_list.append(title.text)
```
